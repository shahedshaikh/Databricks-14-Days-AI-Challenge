```python
# Create sample data
data = [
    (101, "Alex", "Data Analyst", 60000),
    (102, "Riya", "Data Scientist", 85000),
    (103, "John", "Data Engineer", 90000),
    (104, "Sara", "Data Analyst", 65000),
    (105, "Mike", "ML Engineer", 95000)
]

columns = ["emp_id", "name", "role", "salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show data
df.show()
```

Select specific columns

```python
df.select("name", "role").show()
```

Filter data

```python
df.filter(df.role == "Data Analyst").show()
```

Filter with condition

```python
df.filter(df.salary > 80000).show()
```

Group By and aggregate

```python
df.groupBy("role").count().show()
```

Sort data

```python
df.orderBy(df.salary.desc()).show()
```
