UDF â€“ Create Derived Feature (Age Group)

```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import udf

def age_group(age):
    if age < 25:
        return "Young"
    elif age < 40:
        return "Adult"
    else:
        return "Senior"

age_udf = udf(age_group, StringType())

df = df.withColumn(
    "Age_Group",
    age_udf(F.col("Age"))
)

df.select("Age", "Age_Group").display()
```
