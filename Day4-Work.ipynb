```python
%sql
CREATE CATALOG IF NOT EXISTS main;
CREATE SCHEMA IF NOT EXISTS main.default;

CREATE VOLUME IF NOT EXISTS main.default.sales_volume;
```

```python
df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv("/Volumes/main/default/sales_volume/sales.csv")
)
```

```python
from pyspark.sql.functions import col

clean_df = df.select(
    col("Order No").alias("order_no"),
    col("Item ID").alias("item_id"),
    col("Sales Date").alias("sales_date"),
    col("Customer Id").alias("customer_id"),
    col("City Id").alias("city_id"),
    col("Discount Percent").alias("discount_percent")
)
```

```python
clean_df.write.format("delta") \
  .mode("overwrite") \
  .save("/Volumes/main/default/sales_volume/delta/ecommerce_sales")
```

```python
spark.read.format("delta") \
  .load("/Volumes/main/default/sales_volume/delta/ecommerce_sales") \
  .show()
```
